```python
from pathlib import Path # This is a modern Python way to work with Paths so you need less strings. Also see: https://realpython.com/python-pathlib/

Path.cwd() # Show current working directory
```

```python
base_dir = Path("Universe Data") # Use Path to define base directory for Data based on the current working directory

file_path = max(  # Get the most recently modified file
    base_dir.glob("P123_Screen_*.csv"),   #The glob module finds pathnames using pattern matching rules similar to the Unix shell - See also: https://docs.python.org/3/library/glob.html
    key=lambda p: p.stat().st_mtime  # Get the most recently modified file
)

with file_path.open("r", encoding="utf-8-sig") as f: # Open the file (.open) - with statement makes sure file will be closed again - See also: https://realpython.com/read-write-files-python/ - "r" is for read only access
    lines = [line.rstrip("\n") for line in f] # Read all lines and strip newline characters / https://realpython.com/python-strip/

def clean_metadata_line(line):
    # Split CSV row and return first non-empty cell
    parts = [p.strip() for p in line.split(",")]
    for p in parts:
        if p:
            return p
    return None

metadata = {
    "title": clean_metadata_line(lines[0]) if len(lines) > 0 else None,
    "date": clean_metadata_line(lines[1]) if len(lines) > 1 else None,
    "notes": clean_metadata_line(lines[2]) if len(lines) > 2 else None,
    "file_name": file_path.name,
}

metadata

```

```python


df = pd.read_csv(file_path, skiprows=3) # Skip first 3 rows which contain metadata and read data into dataframe

# Drop trailing empty column caused by last comma
df = df.loc[:, ~df.columns.str.startswith("Unnamed")] # Drop unnamed columns / https://www.geeksforgeeks.org/pandas/how-to-drop-unnamed-column-in-pandas-dataframe/

df.tail(20)
```
